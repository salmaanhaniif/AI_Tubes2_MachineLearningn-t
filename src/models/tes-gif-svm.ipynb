```python
import numpy as np
from SVM import BinarySVM 
from sklearn.datasets import make_blobs

# TOY DATA (Data Mainan) 2 Dimensi
X_demo, y_demo = make_blobs(n_samples=50, centers=2, n_features=2, random_state=42, cluster_std=2.5)
# Ubah label dari {0,1} ke {-1,1}
y_demo = np.where(y_demo == 0, -1, 1)

# INISIALISASI MODEL BONUS
print("Sedang melatih model demo...")
svm_bonus = BinarySVM(kernel='linear', C=1.0, max_iter=50)
svm_bonus.fit(X_demo, y_demo)
svm_bonus.generate_training_gif(X_demo, y_demo, output_path='svm_bonus.gif', fps=5)
```
